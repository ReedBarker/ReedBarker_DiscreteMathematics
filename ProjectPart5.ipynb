###(b) Visualize your structure using graphviz package.

```python
import graphviz

diG = graphviz.Digraph(comment='Treasure Map')

# Vertices
diG.node('Starting Point')
diG.node('Beach')
diG.node('Forest')
diG.node('Mountain')
diG.node('Treasure Chest')

# Edges with weights
diG.edge('Starting Point', 'Beach', '2')
diG.edge('Starting Point', 'Forest', '4', constraint='false')
diG.edge('Beach', 'Forest', '1', constraint='false')
diG.edge('Forest', 'Beach', '1', constraint='false')
diG.edge('Beach', 'Mountain', '5', constraint='false')
diG.edge('Forest', 'Mountain', '3')
diG.edge('Mountain', 'Treasure Chest', '2')
diG
```
