##1. Define a function in python and graph it.

I found a really cool, function called the Archimedean Spiral that I am excited to share. The Archimedean Spiral is exactly what it sounds like, a fun spiral but the more you change values, the more crazy it can get. The function is as follows: 

####r=a+bθ

This function has three features a, b, and θ. If you want to play with it on desmos, you have to type out "theta"

In the following code, there is a very basic program that makes a basic spiral. theta (θ) is created using numpy's linspace to create the actual line by having a start point and an end point.

```python
import numpy as np
from matplotlib import animation,pyplot as plt


theta = np.linspace(0, 10 * np.pi, 600) #(line start, line end, "resolution") creates the line
a = 1 #constant
b = 5 #constant
radius = a + b * theta
x = radius * np.cos(theta)
y = radius * np.sin(theta)

plt.plot(x, y)
plt.title('"Basic" Archimedean Spiral')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Tweaking some numbers here and there can make things go a little wild.

```python
import numpy as np
from matplotlib import animation,pyplot as plt


theta = np.linspace(-1, 63000 * np.pi, 600) #(start, end, "resolution")
a = 170
b = -0.1
radius = a + b * theta
x = radius * np.cos(theta)
y = radius * np.sin(theta)

plt.plot(x, y)
plt.title('Flowering Archimedean Spiral')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
